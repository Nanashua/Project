###Jumlah keseluruhan kota

```python
# Group data berdasarkan kolom 'country' dan 'state', kemudian hitung profit tiap kota
profit_per_city = df.groupby(['country', 'state'])['profit'].sum().reset_index()

# Tampilkan 10 data teratas
print(profit_per_city.head())

# Visualisasikan data menggunakan bar plot
plt.figure(figsize=(12, 6))
plt.bar(profit_per_city['state'], profit_per_city['profit'])
plt.xlabel('State')
plt.ylabel('Total Profit')
plt.title('Total Profit per State')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Kota bagian mana yang paling sering terjadi transaksi

---


```python
# Group data berdasarkan kolom 'quantity'dan 'state', kemudian hitung transaksi tiap kota
quantity_per_city = df.groupby(['state'])['quantity'].sum().reset_index()

quantity_per_city_sorted = quantity_per_city.sort_values(by='quantity', ascending=False)

# Tampilkan 10 data teratas
print(quantity_per_city_sorted.head())
```

```python
# Buat plot menggunakan seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=quantity_per_city_sorted.head(10), x='state', y='quantity', palette='viridis')
plt.xlabel('State')
plt.ylabel('Total Quantity')
plt.title('Total Quantity per State (Top 10)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Tampilkan plot
plt.show()
```
